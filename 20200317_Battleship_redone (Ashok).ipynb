### Battleship, One Player

```python
import random

rc_num = 4
board = [['0'] * rc_num for _ in range(rc_num)]

def print_board(board):
    for row in board:
        print (" ".join(row))
        
def random_row(board):
    return random.randrange(0, len(board))

def random_col(board):
    return random.randrange(0, len(board))

ship_row_1 = random_row(board)
ship_col_1 = random_col(board)

ship_row_2 = ship_row_1
ship_col_2 = ship_col_1

while (ship_row_2==ship_row_1) and (ship_col_2==ship_col_1):
    ship_row_2 = random_row(board)
    ship_col_2 = random_col(board)

hit_count = 0

for turn in range(3):
    guess_row = int(input("\nGuess row: enter a number between 0 and 3: "))
    guess_col = int(input("Guess column: enter a number between 0 and 3: "))
    
    if guess_row < 0 or guess_row > (rc_num-1) or guess_col < 0 or guess_col > (rc_num-1):
        print("Oops, that landed on land")
    
    elif board[guess_row][guess_col] == '*' or board[guess_row][guess_col] == 'x':
        print('You already guessed that one')
                    
    elif (guess_row == ship_row_1 and guess_col == ship_col_1) or (guess_row == ship_row_2 and guess_col == ship_col_2):
        board[guess_row][guess_col] = '*'
        hit_count +1
        print ("\nCongratulations!, first ship down ")
                    
        if hit_count == 2:
            print("\nCongratulations, second ship down")
            print(board)
            break
    else:
        board[guess_row][guess_col] = "x"
        print('\nYou missed!')
    
    print(f"\nYou finished {turn+1} turns\n")
    print('This is how your board looks')
    print_board(board)

print("\nGame over!")
```

### Two Player

```python
import random

rc_num = 4
board = [['0'] * 4 for _ in range(rc_num)]

print("Let's play Battleship!")
print("This is a 2 player game")

player_1 = input("Enter first name: ")
player_2 = input("Enter second name: ")
players = [player_1, player_2]


def print_board(board):
    for row in board:
        print (" ".join(row))
        
def random_row(board):
    return random.randrange(0, len(board))

def random_col(board):
    return random.randrange(0, len(board))

def random_player(players):
    return random.choice(players)


ship_row_1 = random_row(board)
ship_col_1 = random_col(board)

ship_row_2 = ship_row_1
ship_col_2 = ship_col_1

while ship_row_1 == ship_row_2 and ship_col_1 == ship_col_2:
    ship_row_2 = random_row(board)
    ship_col_2 = random_row(board)
    

player_start = random_player(players)
if player_start == player_1:
    print(player_1, "starts the game.")
else:
    print(player_2, "starts the game.")    

hit_count = 0

for turn in range(3):
    print(f"{player_start} plays")
    guess_row = int(input("\nGuess row: enter a number between 0 and 3: "))
    guess_col = int(input("Guess column: enter a number between 0 and 3: "))

    if guess_row < 0 or guess_row > len(board)-1 or guess_col < 0 or guess_col > len(board)-1:
        print("Oops, that landed on land")

    elif board[guess_row][guess_col] == '*' or board[guess_row][guess_col] == 'x':
        print('You already guessed that one')

    elif guess_row == ship_row_1 and guess_col == ship_col_1 or guess_row == ship_row_2 and guess_col == ship_col_2:
        board[guess_row][guess_col] = '*'
        hit_count +1
        print ("\nCongratulations! ")

        if hit_count == 2:
            print("\nCongratulations, second ship down")
            print(board)
            break

    else:
        board[guess_row][guess_col] = "x"
        print('\nYou missed!')

    print(f"\nYou finished {turn+1} turns\n")
    print('This is how your board looks')
    print_board(board)

print("\nGame over!", player_start)
```

### (One Player) Where input number ranges from 1 to 4 (both inclusive)

```python
import random

rc_num = 4
board = [['0'] * rc_num for _ in range(rc_num)]

def print_board(board):
    for row in board:
        print (" ".join(row))
        
def random_row(board):
    return random.randrange(1, len(board))

def random_col(board):
    return random.randrange(1, len(board))

ship_row_1 = random_row(board)
ship_col_1 = random_col(board)

ship_row_2 = ship_row_1
ship_col_2 = ship_col_1

while ship_row_2==ship_row_1: 
    ship_row_2 = random_row(board)
    
while ship_col_2==ship_col_1:
    ship_col_2 = random_col(board)

    
hit_count = 0


for turn in range(3):
    guess_row = int(input("\nGuess row: enter a number between 1 and 4: "))
    guess_col = int(input("Guess column: enter a number between 1 and 4: "))
    
    if guess_row < 1 or guess_row > (rc_num) or guess_col < 0 or guess_col > (rc_num):
        print("Oops, that's not even in the ocean")
    
    elif board[guess_row-1][guess_col-1] == '*' or board[guess_row-1][guess_col-1] == 'x':
        print('You already guessed that one')
                    
    elif (guess_row == ship_row_1 and guess_col == ship_col_1) or (guess_row == ship_row_2 and guess_col == ship_col_2):
        board[guess_row-1][guess_col-1] = '*'
        hit_count +1
        print ("\nCongratulations!, first ship down ")
                    
        if hit_count == 2:
            print("\nCongratulations, second ship down")
            print(board)
            break
    else:
        board[guess_row-1][guess_col-1] = "x"
        print('\nYou missed!')
    
    print(f"\nYou finished {turn+1} turns\n")
    print('This is how your board looks')
    print_board(board)


print("\nGame over!")
```
